## Create 2 separate csv files having male and female shopper information

```python
dat_group=dat.groupby('Gender')   #Group Data based on Gender
dat_male=dat_group.get_group('Male')    #Get data for Males
dat_female=dat_group.get_group('Female')    #Get data for Females
dat_male.to_csv('Male_Dat.csv')             #Save the male data information as Male_Dat.csv
dat_female.to_csv('Female_Dat.csv')             #Save the male data information as female_Dat.csv
```

## Create a Pie Chart for Male and Female shoppers showing their age distribution

```python
male_age=list(dat_male['Age'])              #Converts a dataframe column into a list
female_age=list(dat_female['Age'])             #Converts a dataframe column into a list
def age_coun(age):                           #A function which calculates frequency of age
    age_val=[]                             # list having age values
    age_coun=[]                           # List having the corresponding age frequency
    for i in age:                         # A Loop for creating 2 list having age versus number of people
        if i not in age_val:
            s=0
            for j in range(0,len(age)):
                if age[j]==i:
                    s+=1
            age_val.append(i)
            age_coun.append(s)
    return age_val,age_coun
# Returning count of age versus number of people for males
[male_age,male_age_coun]=age_coun(male_age)


# Plot for Male Age
fig=plt.figure(figsize=(10,8))
plt.bar(male_age,male_age_coun, color='maroon', width=0.8)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age distribution of Male Shoppers')
plt.show()
```

```python
# Returning count of age versus number of people for feales
[female_age,female_age_coun]=age_coun(female_age)


# Plot for Male Age
fig=plt.figure(figsize=(10,8))
plt.bar(female_age,female_age_coun, color='maroon', width=0.8)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age distribution of Female Shoppers')
plt.show()
```

From the age plots we can infer the range of ages for male and female shoppers are almost same between (18-70) for males and (18-68) for females.The maximum shoppers from males belong to ages 19, 32, 48 respectively for males and other age groups are on the lower side. For Females it is highly variable and maximum shoppers belong to age 23, 31, 32, 35, 49 respectively. The number of female shoppers are more than male shoppers.

## Plot Histograms of Male and Female annual income and spending score

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12,14)) #Initialization of plot commands
# Male Annual Income Plots
axes[0][0].hist(dat_male['Annual Income (k$)'],bins=10,histtype='bar',rwidth=0.5,align='mid')
axes[0][0].set_xlabel('Annual Income Range (k$)', fontsize=10)
axes[0][0].set_ylabel('Frequency of Annual Income', fontsize=10)
axes[0][0].set_title('Histogram for Male Annual Income')

# Male Spending Score Plots
axes[0][1].hist(dat_male['Spending Score (1-100)'],bins=10,histtype='bar',rwidth=0.5,align='mid')
axes[0][1].set_xlabel('Spending Score', fontsize=10)
axes[0][1].set_ylabel('Frequency of Spending Score', fontsize=10)
axes[0][1].set_title('Histogram for Male Spending Score')


# Female Annual Income Plots
axes[1][0].hist(dat_female['Annual Income (k$)'],bins=10,histtype='bar',rwidth=0.5,align='mid')
axes[1][0].set_xlabel('Annual Income Range (k$)', fontsize=10)
axes[1][0].set_ylabel('Frequency of Annual Income', fontsize=10)
axes[1][0].set_title('Histogram for Female Annual Income')

# Female Spending Score Plots
axes[1][1].hist(dat_female['Spending Score (1-100)'],bins=10,histtype='bar',rwidth=0.5,align='mid')
axes[1][1].set_xlabel('Spending Score', fontsize=10)
axes[1][1].set_ylabel('Frequency of Spending Score', fontsize=10)
axes[1][1].set_title('Histogram for Female Spending Score')
```

The Annual Income Histograms indicate that frequency of higher income people are more in females as compared with males. The highest two income groups in males is between (50,000-60,000 USD) and (80,000-90,000 USD) whereas for females it is (60,000-70,000 USD) and (70,000-80,000 USD) respectively.
The spending scores indicate that the low spending scores are more for males as compared to females and the high spending scores are more for females as compared to males.

## Explore the relationship between age and annual income for male and female shoppers separately

```python
dat_male=dat_male.sort_values(by='Age')    #Sorting data frame values in accordance with age in ascending order by default
dat_female=dat_female.sort_values(by='Age')
mal_age=list(dat_male['Age']) # Sorted Age
mal_sal=list(dat_male['Annual Income (k$)'])   #Age sorted Salary
female_age=list(dat_female['Age']) # Sorted Age
female_sal=list(dat_female['Annual Income (k$)'])   #Age sorted Salary            
```

```python
def plots(g_age,g_sal):      # A Function which accepts age and salary sorted according to ascending order of age
    sal_g=[]
    age=[]
    sal=[]
    for i in range(0,len(g_age)):
    
        if i==0:
            age.append(g_age[i])
            sal.append(g_sal[i])
        else:
            if g_age[i] in age:
                sal.append(g_sal[i])
            else:
                age.append(g_age[i])
                sal_g.append(sal)
                sal=[]
                sal.append(g_sal[i])
        if i==len(g_age)-1:
            sal_g.append(sal)
        
    fig=plt.figure(figsize=(20,12))
    plt.boxplot(sal_g,positions=age)
    plt.title("Age wise salary distribution")
    plt.xlabel('Age')
    plt.ylabel('Annual Salary (k$)')
    plt.show()
```

```python
plots(female_age,female_sal)    # Plot for Female Shoppers
```

The relationship between Age and Annual Salary for Female shoppers is displayed as a boxplot. The red horizontal line indicates the median salary and for cases with only 1 data point the box plot cannot be generated. For more information kindly read about box plots.

As such a meaningful inference cannot be drawn based on the results for Female shoppers as there is a wide gap in income within same age, but in general between the age groups of 27-34 and 44-47 the median income is high.

```python
plots(mal_age,mal_sal)    # Plot for Male Shoppers
```

For males the data suggests the salary gap is also quiet large within the same age group. But for age groups 27-34 the median income is the highest.

## Explore the relationship between age and Spending Score for male and female shoppers separately

```python
def plots_1(g_age,g_sal):      # Same above function but changes only in labels and title as other functions as same
    sal_g=[]
    age=[]
    sal=[]
    for i in range(0,len(g_age)):
    
        if i==0:
            age.append(g_age[i])
            sal.append(g_sal[i])
        else:
            if g_age[i] in age:
                sal.append(g_sal[i])
            else:
                age.append(g_age[i])
                sal_g.append(sal)
                sal=[]
                sal.append(g_sal[i])
        if i==len(g_age)-1:
            sal_g.append(sal)
        
    fig=plt.figure(figsize=(20,12))
    plt.boxplot(sal_g,positions=age)
    plt.title("Age wise Spending Score")
    plt.xlabel('Age')
    plt.ylabel('Spending Score')
    plt.show()
```

```python
plots_1(female_age,female_spd)    # Plot for Female Shoppers
```

There is large variation in spending patterns within same age groups, but the highest median spending score is amongst shoppers less than the age of 36. They can be split as age group 18-24 and 28-33.

```python
plots_1(mal_age,mal_spd)    # Plot for Male Shoppers
```

The mean spending score seem less than female counterparts for age groups less than 24 and then begin to increase. Maximum score occurs between age group of 24-32. Beyond the age of 40 though the spending score remains stable and less than what is was for 40 years and younger.

## Relationship between Spend Score and Annual Salary for Male and Female Shoppers

```python
dat_male=dat_male.sort_values(by='Annual Income (k$)')#Sorting data frame values in accordance with Income in ascending order by default
dat_female=dat_female.sort_values(by='Annual Income (k$)')
mal_inc=list(dat_male['Annual Income (k$)']) # Sorted Income
mal_spd=list(dat_male['Spending Score (1-100)'])   #Income sorted Spend score
female_inc=list(dat_female['Annual Income (k$)']) # Sorted Income
female_spd=list(dat_female['Spending Score (1-100)'])   #Income sorted Spend score
```

```python
fig=plt.figure(figsize=(10,8))   #For female shoppers
plt.plot(female_inc,female_spd,'o-')
plt.title('Income vs Spending Score: Female Shoppers')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
fig=plt.figure(figsize=(10,8))   #For male shoppers
plt.plot(mal_inc,mal_spd,'o-')
plt.title('Income vs Spending Score: Male Shoppers')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```

The spending scores are highly variable with respect to Annual Income. But for salary groups between 40-70 '000 USD the spending score variation are relatively similar with less variation and on the higher side for both age groups. Beyond the 70 '000 USD there are more points in males having low spending scores as compared with female shoppers.
